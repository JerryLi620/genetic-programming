```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the equation function
def equation(x):
    return ((2 + ((-2 - 2) * (-3 + x))) + ((x - 2) * x))

# Generate x values for the plot
x_values = np.linspace(-10, 10, 400)  # Adjust the range as needed
y_values = [equation(x) for x in x_values]

# Load the CSV data
data = pd.read_csv('dataset1.csv')
x_data = data['x']
y_data = data['f(x)']

# Plot the equation
plt.plot(x_values, y_values, label='Model (Given Equation)', color='blue')

# Plot the CSV data
plt.scatter(x_data, y_data, color='red', label='Data from CSV')

# Set up the plot aesthetics
plt.title('Visualization of Model vs Data')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
